```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import folium as fl
```

```python
dati = pd.read_csv(r'C:\Users\glocc\Downloads\COMUNI_52.csv')
```

```python
dati.head()
```

```python
coordinate = pd.read_excel(r'C:\Users\glocc\Downloads\italy_geo.xlsx')
```

```python
coordinate.info()
```

```python
dati.columns = dati.columns.str.lower()
coordinate.columns = coordinate.columns.str.lower()

merged = pd.merge(dati, coordinate, on='comune', how='inner')
```

```python
merged.info()
```

```python
def format_italian_number(n):
    return f"{n:,}".replace(",", ".")

print(format_italian_number(9000000))
```

```python
# Crea la mappa centrata sull'Italia
mappa = fl.Map(location=[42.5, 12.5], zoom_start=6)

# Aggiungi un cerchio per ogni comune
for _, row in merged.iterrows():
    fl.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=row['tot_visitatori'] / 1800000,  # Adatta il divisore in base alla scala dei tuoi dati
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.4,
        popup=f"{row['comune']}: {format_italian_number(row['tot_visitatori'])} visitatori"
    ).add_to(mappa)

# Mostra la mappa
mappa
```

```python
mappa.save("mappa_comuni.html")
```
